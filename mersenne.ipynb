```python
import math
import matplotlib.pyplot as plt
import mpmath
import numpy as np
from numpy.polynomial import Polynomial

START = 521 # first Mersenne exponent used for the graphs

mersenne_exponents = [
    2,
    3,
    5,
    7,
    13,
    17,
    19,
    31,
    61,
    89,
    107,
    127,
    521,
    607,
    1279,
    2203,
    2281,
    3217,
    4253,
    4423,
    9689,
    9941,
    11213,
    19937,
    21701,
    23209,
    44497,
    86243,
    110503,
    132049,
    216091,
    756839,
    859433,
    1257787,
    1398269,
    2976221,
    3021377,
    6972593,
    13466917,
    20996011,
    24036583,
    25964951,
    30402457,
    32582657,
    37156667,
    42643801,
    43112609,
    57885161,
    74207281,
    77232917,
    82589933,
    136279841,
]

%matplotlib inline

count = 0
xvals = []
yvals = []
for count, exponent in enumerate(mersenne_exponents, start=1):
    if exponent >= START:
        xvals.append(math.log2(exponent))
        yvals.append(count)

fit = Polynomial.fit(xvals, yvals, deg=2)
print(fit)
x = np.linspace(xvals[0], xvals[-1], 100)

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(xvals, yvals, 'ro')
ax.plot(x, fit(x))
ax.grid()
ax.set_title("Mersenne prime exponent count")
ax.set_xlabel("$log_{2}$(exponent)")
ax.set_ylabel("count")

count = 0
xvals = []
yvals = []
for count, exponent in enumerate(mersenne_exponents, start=1):
    if exponent >= START:
        xvals.append(math.log2(mpmath.li(exponent)))
        yvals.append(count)

fit = Polynomial.fit(xvals, yvals, deg=2)
print(fit)
x = np.linspace(xvals[0], xvals[-1], 100)

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(xvals, yvals, 'ro')
ax.plot(x, fit(x))
ax.grid()
ax.set_title("Mersenne prime exponent count")
ax.set_xlabel("$log_{2}$(li(exponent))")
ax.set_ylabel("count")
```
